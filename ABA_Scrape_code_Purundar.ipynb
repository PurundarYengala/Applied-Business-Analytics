```python
import requests
import pandas as pd
import os

def fetch_data(api_key):
    url = 'https://data.cityofnewyork.us/resource/nc67-uf89.json'
    params = {
        '$$app_token': api_key,
        '$limit': 50000  
    }
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        return response.json()
    else:
        raise Exception("Failed to fetch data: HTTP Status Code {}".format(response.status_code))

def save_to_csv(data, directory):
    if not os.path.exists(directory):
        os.makedirs(directory)  # Create the directory if it doesn't exist
    filepath = os.path.join(directory, 'nyc_parking_violations.csv')
    df = pd.DataFrame(data)
    df.to_csv(filepath, index=False)
    print(f"Data saved to '{filepath}'.")

def main():
    api_key = 'GzZFsKlM9wPeuIdjb4vzLZpX7'  
    directory = r'C:\Users\Purundar\Downloads\Open parking and camera violations\CSV'  # Specified directory path
    data = fetch_data(api_key)
    save_to_csv(data, directory)

if __name__ == '__main__':
    main()
```
